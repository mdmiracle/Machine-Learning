```python
#x=df.loc[:,['left','salary']]
input=df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
input
```

```python
dummies=pd.get_dummies(input.salary)
input_with_dummies=pd.concat([input,dummies],axis=1)
input_with_dummies.drop(['salary'],axis=1,inplace=True)
input_with_dummies.head()
```

```python
label= df.left
label
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_with_dummies,label,test_size=0.2)
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
