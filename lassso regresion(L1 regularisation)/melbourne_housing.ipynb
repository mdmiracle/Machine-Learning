```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
len(cols)
```

```python
data=df[cols]
data.head()
```

```python
data.isna().sum()
```

```python
data.Car=data.Car.fillna(0)
data[data.Car==0]
```

```python
data.BuildingArea=data.BuildingArea.fillna(data.BuildingArea.mean())
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data=pd.get_dummies(data,drop_first=True)
data.shape
input=data.drop(['Price'],axis=1)
label=data.Price
input.head()
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,label_train,label_test=train_test_split(input,label,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(input_train,label_train)
```

```python
reg.score(input_train,label_train)
```

```python
reg.score(input_test,label_test)
```

```python
from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=100,max_iter=1000,tol=0.1)
lasso_reg.fit(input_train,label_train)
```

```python
lasso_reg.score(input_train,label_train)
```

```python
lasso_reg.score(input_test,label_test)
```

```python
reg.coef_
```

```python
lasso_reg.coef_
```

```python
ar1=np.array([1,2,3,4])
ar2=np.array([3,4,5,6])
(ar1==ar2).sum()
(lasso_reg.coef_<=0.00001).sum()
```

```python
lasso_reg.n_features_in_,lasso_reg.n_iter_,lasso_reg.intercept_,reg.intercept_
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(input_train,label_train)
```

```python
ridge_reg.score(input_train,label_train)
```

```python
ridge_reg.score(input_test,label_test)
```

```python
(ridge_reg.coef_<=0.001).sum()
```
